```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle

with open('datapoints_dict.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
apls = []

for key in data:
    if not np.isnan(data[key]['apl']):
        apls.append(data[key]['apl'])
        
apls = np.array(apls)
```

```python
plt.hist(apls)
plt.xlabel('Average APL (Ã…$^2$)')
plt.ylabel('Amount')
plt.title('Average APL histogram of 714 simulations')
plt.show()
```
