```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import os

df = pd.read_json('simulations.json')
df
```

```python
full_df = df.dropna(axis = 1)
full_df
```

```python
index = np.arange(0,1,0.001)
modified_FF = pd.DataFrame(index=range(1000),columns=[full_df.columns])
modified_FF.index = index

for i in full_df.columns:
    values = []
    for pair in full_df[i]['FormFactor']:
        values.append(round(pair[1],5))
    modified_FF[i] = values
    
modified_FF
```

```python
index = np.arange(1,196,1)

names = []
for i in full_df.columns:
    names.append(str(i) + 'x')
    names.append(str(i) + 'y')
    
modified_TD = pd.DataFrame(index=range(195),columns=[names])
modified_TD.index = index

for i in full_df.columns:
    x_values = []
    y_values = []
    for pair in full_df[i]['TotalDensity']:
        x_values.append(round(pair[0],5))
        y_values.append(round(pair[1],5))
    #print(len(x_values))
    
    modified_TD[str(i) + 'x'] = pd.Series(x_values)
    modified_TD[str(i) + 'y'] = pd.Series(y_values)

modified_TD

#full_df[1]['TotalDensity']
```

```python
modified_FF = (modified_FF-modified_FF.mean())/modified_FF.std()

for i in modified_FF.columns:
    plt.plot(np.arange(0,1,0.001), modified_FF[i])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scaled FormFactor of 609 simulations')
plt.show()
```

```python
from sklearn.cluster import KMeans
import sklearn.metrics as sm

model = KMeans(n_clusters=10)
model.fit(modified_FF.values)
```

```python
colormap = np.array(['pink','purple','blue','red','black','orange', 'cyan', 'green', 'violet', 'grey'])

for i in modified_FF.columns:
    plt.scatter(np.arange(0,1,0.001), modified_FF[i],c=colormap[model.labels_], s=40)
    
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scaled FormFactor of 609 simulations with K Mean Classification')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(modified_FF.values)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.title('PCA for FormFactor of 609 simulations')
plt.show()
```

```python
pca = PCA(n_components=2).fit_transform(modified_FF.values)
principalDf = pd.DataFrame(pca, columns = ['principal component 1', 'principal component 2'])
plt.scatter(principalDf['principal component 1'],principalDf['principal component 2'],s=30,c='goldenrod',alpha=0.5)
plt.title('PCA scatterplot for FormFactor of 609 simulations')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
model = KMeans(n_clusters=2)
model.fit(principalDf)
colormap = np.array(['blue','red'])
plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'],c=colormap[model.labels_], s=40)
plt.title('PCA FormFactor of 609 simulations witk K means')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```
