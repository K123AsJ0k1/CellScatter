```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('simulations.json')
df
```

```python
full_df = df.dropna(axis = 1)
full_df
```

```python
from sklearn.preprocessing import StandardScaler

index = np.arange(0,1,0.001)
modified_FF = pd.DataFrame(index=range(1000),columns=[full_df.columns])
modified_FF.index = index

for i in full_df.columns:
    values = []
    for pair in full_df[i]['FormFactor']:
        values.append(round(pair[1],5))
    modified_FF[i] = values
    
modified_FF = (modified_FF-modified_FF.mean())/modified_FF.std()
#modified_FF = StandardScaler().fit_transform(modified_FF)
modified_FF
```

```python
index = np.arange(1,196,1)

names = []
for i in full_df.columns:
    names.append(str(i) + 'x')
    names.append(str(i) + 'y')
    
modified_TD = pd.DataFrame(index=range(195),columns=[names])
modified_TD.index = index

for i in full_df.columns:
    x_values = []
    y_values = []
    for pair in full_df[i]['TotalDensity']:
        x_values.append(round(pair[0],5))
        y_values.append(round(pair[1],5))
    
    modified_TD[str(i) + 'x'] = pd.Series(x_values)
    modified_TD[str(i) + 'y'] = pd.Series(y_values)

modified_TD
```

```python
import random

names = []
for i in input_data.columns:
    names.append(i[0])

train = random.sample(names,64)
test = []
for number in names:
    if not number in train:
        test.append(number)

print(train)
print(test)
```

```python
train_input_values = []
for name in train:
    train_input_values.append(input_data[name].values)

train_output_values = []
for name in train:
    list_1 = output_data[str(name) + 'x'].values
    list_2 = output_data[str(name) + 'y'].values
    train_output_values.append(np.concatenate((list_1,list_2)))
    
test_input_values = []
for name in test:
    test_input_values.append(input_data[name].values)
    
test_output_values = []
for name in test:
    list_1 = output_data[str(name) + 'x'].values
    list_2 = output_data[str(name) + 'y'].values
    test_output_values.append(np.concatenate((list_1,list_2)))
    
train_input_values = np.array(train_input_values)
train_output_values = np.array(train_output_values)
test_input_values = np.array(test_input_values)
test_output_values = np.array(test_output_values)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

inputs = keras.Input(shape=(1000,))
#dense = layers.Dense(64, activation="relu")
x = layers.Dense(128, activation="relu")(inputs)
x = layers.Dropout(0.15, noise_shape = None, seed = 42)(x)
x = layers.Dense(128, activation="relu")(x)
#x = layers.Dense(128, activation="relu")(x)
#x = layers.Dropout(0.2, noise_shape = None, seed = None)(x)
#x = layers.Dense(128, activation="relu")(x)
#x# = layers.Dense(128, activation="relu")(x)
#x = layers.Dense(128, activation="relu")(x)
#dense_2 = layers.Dense(64, activation="relu")
outputs = layers.Dense(370)(x)
model = keras.Model(inputs=inputs, outputs=outputs)

model.compile(optimizer='adam', loss='mean_squared_error')

history = model.fit(train_input_values, train_output_values, batch_size=16, epochs=20, validation_split=0.2)
test_scores = model.evaluate(test_input_values, test_output_values, verbose=0)
#print(test_scores)
```
